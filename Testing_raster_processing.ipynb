Libraries used

```python
import numpy as np
from osgeo import gdal
import matplotlib.pyplot as plt
```

Defining a function to import and view the raster bands

```python
def import_and_plot(spatial_res, band_index):
    
    """Imports the raster stack using gdal
       Then convert a selected band into a numpy array
       Then plots the 1-band image
       
       Requires:
           The file path for the raster stack
           The index value of the desired raster band
           
       Outputs:
           An image of the i-th band of the raster stack"""
    
    file_path = '/data/Images/Denver/Denver' + str(spatial_res) + '.tif'
    print(file_path)
    # Import
    raster_stack = gdal.Open(file_path)
    # Convert
    raster_array = raster_stack.GetRasterBand(band_index).ReadAsArray()
    # Plot
    plt.imshow(raster_array, cmap = 'viridis');
```

# 10m Rasters

```python
import_and_plot(10, 1)
```

```python
import_and_plot(10, 2)
```
